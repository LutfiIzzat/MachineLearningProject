# Feature selection
Using chi-square analysis, we reduce the dataset by finding features that are most likely to be correlated with our target of cats or dogs

```python
df.columns
```

```python
target = 'Type'

X = df.drop(columns=target)
y = df[target]

display(X.shape)
display(y.shape)
```

```python
from scipy.stats import chi2_contingency
```

```python
# Empty lists to store results
columns = []
chi2_stats = []
p_values = []

# chi square for all cols
for col in X.columns:
    # Convert the column in X to a 1-dimensional array
    col_data = X[col].values.flatten()
            
    # Convert the target column to a 1-dimensional array
    target_data = y.values.flatten()

    # Create contingency table
    ct_table_ind = pd.crosstab(col_data, target_data)
            
    # Compute the chi2 stat and get the value
    c_stat, p, dof, expected = chi2_contingency(ct_table_ind)
            
    # Append results to lists
    columns.append(col)
    chi2_stats.append(c_stat)
    p_values.append(p)

# Create a DataFrame from the results
results_df = pd.DataFrame({'Column': columns, 'Chi2 Statistic': chi2_stats, 'P-Value': p_values})
```

```python
# <= 0.05 = has association, is dependent, rejects null hypothesis

chi = results_df[results_df['P-Value'] <= 0.05].sort_values(by='P-Value', ascending=False).reset_index(drop=True)

chi['P-Value Decimal'] = chi['P-Value'].apply(lambda x: format(x, '.10f'))

print('Relevant features')
display(chi.shape)
display(chi)
```

```python
display(results_df[(results_df['P-Value'] > 0.05)].sort_values(by='P-Value', ascending=False).reset_index(drop=True))
```

```python
cols = chi['Column'].unique().tolist()

X = X[cols]
X.head()
```

## Linear Regression

Linear regression is a statistical method used to model the relationship between a dependent variable (target variable) and one or more independent variables (predictor variables). It works on the principle of fitting a straight line (or hyperplane in higher dimensions) to the data points to make predictions.

```python
X_train
```

```python
X_test
```

```python
y_train
```

In the case of a classification problem, where the target variable (output) represents categories or classes (e.g., Cat and Dog), linear regression is not the most appropriate choice because:

- Continuous Output: Linear regression predicts continuous numeric values. It assumes that the target variable has a continuous range of values. However, in a classification problem, the target variable represents discrete categories (e.g., Cat or Dog), which are not continuous.

- Assumption Violation: Linear regression assumes that the relationship between the independent and dependent variables is linear. It may not accurately capture the underlying relationship between categorical classes and the predictor variables, leading to poor model performance and erroneous predictions.

To address this issue, we typically encode categorical target variables into numerical values. In binary classification problems (where there are only two classes, such as Cat and Dog), it's common to encode one class as 0 and the other class as 1. This allows us to treat the problem as a regression task, where the model learns to predict the continuous probability of belonging to one class versus the other. 

```python
# Linear regression does not accept categorical target unlike logistic regression and k-means. 
# Therefore, y_train and y_test will be re-encoded

label_map = {"Cat": 0, "Dog": 1}

# Map the categorical labels in y_train and y_test using the dictionary
y_train_encoded = y_train.map(label_map)
y_test_encoded = y_test.map(label_map)

print("y_train_encoded:")
print(y_train_encoded)
print("\ny_test_encoded:")
print(y_test_encoded)
```

```python
print('Training set: ', X_train.shape)
print(y_train_encoded.shape)

print('Testing set: ', X_test.shape)
print(y_test_encoded.shape)
```

Applying linear regression to the encoded training and testing set

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

linreg = LinearRegression()

# Train the Model
linreg.fit(X_train, y_train_encoded)

# Predict on Test Data
y_pred_linreg = linreg.predict(X_test)

# Evaluate the Model
mse = mean_squared_error(y_test_encoded, y_pred_linreg)
print("Mean Squared Error:", mse)

print('Training accuracy: ', linreg.score(X_train, y_train_encoded))
print('Testing accuracy: ', linreg.score(X_test, y_test_encoded))
```

The model is correctly classifying only around 35-34% of the instances in both the training and testing datasets. This means that the model's predictive performance is not much better than random guessing, especially considering that a random guess would typically achieve an accuracy of around 50% in a binary classification problem with balanced classes.

Futhermore the training accuracy being similar to or only slightly higher than the testing accuracy suggests that the model might be either underfitting or overfitting:

- Underfitting: If both training and testing accuracies are low and similar, it indicates that the model is too simple to capture the underlying patterns in the data.

- Overfitting: If the training accuracy is significantly higher than the testing accuracy, it suggests that the model has learned to fit the training data too closely and may not generalize well to unseen data.

In classification tasks, especially if the classes are imbalanced or the task is complex, higher accuracy rates are desirable. A low accuracy indicates that there is room for improvement in the model, either by adjusting the model's complexity, tuning hyperparameters, or exploring different algorithms.

#### Underfitting or Overfitting?

Plot learning curves that show the model's performance (e.g., accuracy or error) on the training set and testing set as a function of the number of training instances. This helps visualize whether the model is converging to a stable performance or if it's still learning.

```python
def plot_learning_curves(estimator, X, y):
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, train_sizes=np.linspace(0.1, 1.0, 10), 
        cv=5, scoring='neg_mean_squared_error')

    train_errors = -train_scores.mean(axis=1)
    test_errors = -test_scores.mean(axis=1)

    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, train_errors, 'o-', label='Training error')
    plt.plot(train_sizes, test_errors, 'o-', label='Testing error')
    plt.xlabel('Training instances')
    plt.ylabel('Mean Squared Error')
    plt.title('Learning Curves')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_learning_curves(linreg, X_train, y_train_encoded)
```

The Newton-Cholesky method is an optimization algorithm used primarily for solving logistic regression problems. It's a variant of the Newton-Raphson optimization algorithm, which is widely used in optimization tasks, particularly in logistic regression.

Newton Cholesky is effective when the number of samples (data points) is much larger than the number of features (attributes or variables). Since The dataset has a lot more datapoints than the categorical features, it theoretically can increase the efficacy of the training and testing accuracy

```python
logreg = LogisticRegression(solver='newton-cholesky', penalty='l2', C=2, max_iter=1000) #initialize logistic regression model with 'l2' penalty
logreg.fit(X_train, y_train) 

print('Training accuracy: ', logreg.score(X_train, y_train))
print('Testing accuracy: ', logreg.score(X_test, y_test))

logcompare.append(['newton-cholesky', logreg.score(X_train, y_train), logreg.score(X_test, y_test)])
```

Lastly, the Liblinear method is used below, using a L2 penalty and a coefficient of 2, This is due to the fact that The solver uses a Coordinate Descent (CD) algorithm that solves optimization problems by  performing approximate minimization along coordinate directions.

A penalty of L2 is used due to the high dimensionality of the selected features, thus favours the logisticRegression Model

```python
logreg = LogisticRegression(solver='liblinear', penalty='l2', C=2, max_iter=1000) # logistic regression model with 'l2' penalty
logreg.fit(X_train, y_train) 

print('Training accuracy: ', logreg.score(X_train, y_train))
print('Testing accuracy: ', logreg.score(X_test, y_test))

logcompare.append(['liblinear', logreg.score(X_train, y_train), logreg.score(X_test, y_test)])
```

## Permutation Importance

Permutation importance is a model-agnostic method for calculating feature importance that can be used with logistic regression.

Permutation importance measures the decrease in model performance that occurs when a feature's values are randomly shuffled after training, thereby severing the feature's relationship to the target variable.

A feature is significant if changing its values causes the model's performance to significantly decline.


```python
from sklearn.inspection import permutation_importance

result = permutation_importance(logreg, X_test, y_test, n_repeats=10, random_state=42)


feature_importance = pd.DataFrame({'Feature': X.columns,
                                   'Importance': result.importances_mean,
                                   'Standard Deviation': result.importances_std})
feature_importance = feature_importance.sort_values('Importance', ascending=True)


ax = feature_importance.plot(x='Feature', y='Importance', kind='barh', figsize=(10, 6), yerr='Standard Deviation', capsize=4)
ax.set_xlabel('Permutation Importance')
ax.set_title('Permutation Importance with Standard Deviation')
```

Based on the graph above, we can see that the breed has the highest importance and will more likely affect the likelihood of adoption of pets

## K-Means

### Finding the Optimal Number of Clusters using Elbow Method

```python
from sklearn.cluster import KMeans
```

```python
# 3. Create and fit the k-means clustering model
kmeans = KMeans(n_clusters=4, random_state=0) 
kmeans.fit(X_train)

# Optionally, you can use the trained model to predict cluster labels for the testing data
cluster_labels = kmeans.predict(X_test)

# 4. Evaluate the performance of the clustering model
# Since k-means is unsupervised, so it won't have traditional evaluation metrics like accuracy
# Evaluate based on inertia or silhouette score.
# Inertia measures how tightly grouped the clusters are. Lower inertia is generally better.
print("Inertia:", kmeans.inertia_)

# List to store inertia values
inertia = []

# Define range of k values to test
k_values = range(1, 11)

# Fit KMeans for each value of k and store inertia
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X_train)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

### Visualizing KMeans Clustering Results on Training and Testing Data (k=2)

```python
# Initialize KMeans with optimal value of k
kmeans = KMeans(n_clusters=2, random_state=0)

# Fit KMeans to the training data
kmeans.fit(X_train)

# Get cluster labels for the training and testing data
train_labels = kmeans.predict(X_train)
test_labels = kmeans.predict(X_test)

# Define colors for each cluster
colors = ['blue', 'green']

# Scatter plot for training data
plt.figure(figsize=(10, 6))
for i in range(2):
    plt.scatter(X_train.values[train_labels == i, 0], X_train.values[train_labels == i, 1], c=colors[i], s=50, alpha=0.5, label=f'Cluster {i}')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=150, label='Centroids')
plt.title('KMeans Clustering - Training Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Scatter plot for testing data
plt.figure(figsize=(10, 6))
for i in range(2):
    plt.scatter(X_test.values[test_labels == i, 0], X_test.values[test_labels == i, 1], c=colors[i], s=50, alpha=0.5, label=f'Cluster {i}')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=150, label='Centroids')
plt.title('KMeans Clustering - Testing Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

### Finding the Optimal Number of Clusters using Silhouette Score

```python
from sklearn.metrics import silhouette_score
```

```python
# 1. Compute silhouette scores for different values of k
silhouette_scores = []
k_values = range(2, 11)  # Adjust the range as needed

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    cluster_labels = kmeans.fit_predict(X_train)
    silhouette_scores.append(silhouette_score(X_train, cluster_labels))

# 2. Plot silhouette scores
plt.plot(k_values, silhouette_scores, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal k')
plt.xticks(k_values)
plt.grid(True)
plt.show()

# 3. Find the optimal k
optimal_k = k_values[np.argmax(silhouette_scores)]
print("Optimal number of clusters (k):", optimal_k)
```

### Visualizing KMeans Clustering Results on Training and Testing Data (k=8)

```python
# Initialize KMeans with optimal value of k
kmeans = KMeans(n_clusters=8, random_state=0)

# Fit KMeans to the training data
kmeans.fit(X_train)

# Get cluster labels for the training and testing data
train_labels = kmeans.predict(X_train)
test_labels = kmeans.predict(X_test)

# Scatter plot for training data
plt.figure(figsize=(10, 6))
for label in range(8):  # Loop through each cluster label
    plt.scatter(X_train.values[train_labels == label, 0], X_train.values[train_labels == label, 1], label=f'Cluster {label}', s=50, alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=150, label='Centroids')
plt.title('KMeans Clustering - Training Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Scatter plot for testing data
plt.figure(figsize=(10, 6))
for label in range(8):  # Loop through each cluster label
    plt.scatter(X_test.values[test_labels == label, 0], X_test.values[test_labels == label, 1], label=f'Cluster {label}', s=50, alpha=0.5)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=150, label='Centroids')
plt.title('KMeans Clustering - Testing Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

The analysis utilized both the elbow method and silhouette score to determine the optimal number of clusters. However, there was a difference in their recommendations: the elbow method suggested two clusters, while the silhouette score pointed towards eight clusters. Despite this difference, both sets of clusters were implemented using KMeans.

Upon visualizing the data, it was observed that although distinct groupings were expected, the large dataset size might have caused some blending or overlap of colors within the same group. This observation applied to both the two-cluster and eight-cluster scenarios.
